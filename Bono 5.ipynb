## Punto 6.2

```python
F=np.arange(10,120,10)
V=np.array((16,45,64,75,70,115,142,167,183,160,221))
e_1=np.array((5,5,5,5,30,5,5,5,5,30,5))
w_1=1/e_1**2
```

```python
def m(w,x,y):
    value=(np.sum(w)*np.sum(w*x*y)-np.sum(w*x)*np.sum(w*y))/(np.sum(w)*np.sum(w*x**2)-
                                                     (np.sum(w*x))**2)
    inc = np.sqrt(np.sum(w)/(np.sum(w)*np.sum(w*x**2)-(np.sum(w*x))**2))
    return [value,inc]
def c(w,x,y):
    value=(np.sum(w*x**2)*np.sum(w*y)-np.sum(w*x)*np.sum(w*x*y))/(np.sum(w)*np.sum(w*x**2)-
                                                     (np.sum(w*x))**2)
    inc=np.sqrt(np.sum(w*x**2)/(np.sum(w)*np.sum(w*x**2)-(np.sum(w*x))**2))
    return [value,inc]
m_1=m(w_1,F,V)
c_1=c(w_1,F,V)
print('m={:.2f}\u00B1{:.2f}'.format(*m_1))
print('c={:.0f}\u00B1{:.0f}'.format(*c_1))
```

## Punto 6.6

```python
x=np.arange(1,11,1)
y=np.array((51,103,150,199,251,303,347,398,452,512))
e_3=np.array((1,1,2,2,3,3,4,5,6,7))
e_4=4*np.ones(10)
e_5=np.array((1,8,8,8,8,8,8,8,8,1))
w_3=1/e_3**2
w_4=1/e_4**2
w_5=1/e_5**2
w=[w_3,w_4,w_5]
```
